```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Import data
df1 = pd.read_csv("C:/Users/u301380/Documents/pyhton2022/SEMVIABLE/beta.csv")
df2 = pd.read_csv("C:/Users/u301380/Documents/pyhton2022/SEMVIABLE/effects.csv")
```

```python
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# Plot the first plot on the first subplot
df1.plot(x="Pathway", y=["β", "f²", "std"], kind="bar", ax=ax1)

ax1.set_xticks(np.arange(len(df1.Pathway)))
ax1.set_xticklabels(df1.Pathway, rotation=90)
ax1.set_xlabel('Pathways')
ax1.set_ylabel('Values')
ax1.set_title('(a) Path Coefficient *p=0.00')


# Plot the second plot on the second subplot
df2.plot(x="Pathways", y=["t-stats"], kind="bar", ax=ax2)
ax2.set_xticks(np.arange(len(df2.Pathways)))
ax2.set_xticklabels(df2.Pathways, rotation=90)
ax2.set_xlabel('Pathways')
ax2.set_ylabel('Values')
ax2.set_title('(b) Specific indirect effects *p=0.00') 
plt.savefig('finalpathways.png', bbox_inches='tight', dpi=300)
plt.savefig('pathwaysfull1.png', bbox_inches='tight', dpi=300)
```
